```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
import os

class DepthMap:
    def _init_(self, showImages):  # Corrected constructor method
        # Load Images
        root = os.getcwd()
        imgLeftPath = os.path.join(root, 'left_image.jpg')
        imgRightPath = os.path.join(root, 'right_image.jpg')
        
        print("Left Image Path:", 'left_image.jpg')  # Check the image path
        print("Right Image Path:", 'right_image.jpg')  # Check the image path

        self.imgLeft = cv.imread(imgLeftPath, cv.IMREAD_GRAYSCALE)
        self.imgRight = cv.imread(imgRightPath, cv.IMREAD_GRAYSCALE)
        
        # Check if images are loaded properly
        if self.imgLeft is None or self.imgRight is None:
            print("Error: One or both images could not be loaded.")
            return

        # Show Images
        if showImages:
            plt.figure()
            plt.subplot(121)
            plt.imshow(self.imgLeft, cmap='gray')
            plt.subplot(122)
            plt.imshow(self.imgRight, cmap='gray')
            plt.show()

    def computeDepthMap(self):
        if self.imgLeft is None or self.imgRight is None:
            print("Error: Images are not loaded properly.")
            return

        nDispFactor = 12
        stereo = cv.StereoBM_create(numDisparities=16 * nDispFactor, blockSize=21)
        disparity = stereo.compute(self.imgLeft, self.imgRight)
        plt.imshow(disparity, 'gray')
        plt.show()

    def computeDepthMapSGBM(self):
        if self.imgLeft is None or self.imgRight is None:
            print("Error: Images are not loaded properly.")
            return

        window_size = 7
        min_disp = 16
        nDispFactor = 14
        num_disp = min_disp * nDispFactor - min_disp

        stereo = cv.StereoSGBM_create(minDisparity=min_disp,
                                      numDisparities=num_disp,
                                      blockSize=window_size,
                                      P1=8 * 3 * window_size ** 2,
                                      P2=32 * 3 * window_size ** 2,
                                      disp12MaxDiff=1,
                                      uniquenessRatio=15,
                                      speckleWindowSize=0,
                                      speckleRange=2,
                                      preFilterCap=63,
                                      mode=cv.STEREO_SGBM_MODE_SGBM_3WAY)

        disparity = stereo.compute(self.imgLeft, self.imgRight).astype(np.float32) / 16.0
        plt.imshow(disparity, 'gray')
        plt.colorbar()
        plt.show()

def demoViewPics():
    dp = DepthMap(showImages=True)

def demoStereoBM():
    dp = DepthMap(showImages=True)
    dp.computeDepthMap('left_image.jpg')

def demoStereoSGBM():
    dp = DepthMap(showImages=True)
    dp.computeDepthMapSGBM('left_image.jpg')

if __name__ == '_main_':
    demoViewPics()
    demoStereoBM()
    demoStereoSGBM()
```
